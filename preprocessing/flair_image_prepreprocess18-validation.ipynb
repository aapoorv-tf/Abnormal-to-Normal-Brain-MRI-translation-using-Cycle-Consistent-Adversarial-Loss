```python
# from medpy.io import load
import nibabel as nib
import matplotlib.pyplot as plt
import random
from sklearn.preprocessing import MinMaxScaler
import glob
import numpy as np
scaler = MinMaxScaler()
```

```python
flair_list = sorted(glob.glob("../dataset/brats2018/BraTS_2018_Data_Validation/*/*flair.nii"))
# mask_list = sorted(glob.glob("../dataset/brats2018/BraTS_2018_Data_Validation/*/*seg.nii"))
```

```python
len(flair_list)
```

```python
for i in range(len(flair_list)):
    flair = nib.load(flair_list[i]).get_fdata()
    flair = scaler.fit_transform(flair.reshape(-1, flair.shape[-1])).reshape(flair.shape)
    for n_slice in range(75,100):
        np.save('../NormalAndAbnormalImages2018/test/flair'+str(i)+'_slice'+str(n_slice)+'.npy', flair[:,:,n_slice])

```

```python
n = 32
slice = 83
flair_img = np.load('../NormalAndAbnormalImages2018/test/flair'+str(n)+'_slice'+str(slice)+'.npy')
```
